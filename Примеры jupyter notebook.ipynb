```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 10, 100)
y = np.cos(x)
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("График косинусоиды")
plt.show()
```
